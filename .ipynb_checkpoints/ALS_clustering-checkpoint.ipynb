```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.manifold import TSNE 
from sklearn.decomposition import PCA

# 解决matplotlib显示中文问题
# 仅适用于Windows
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
data = pd.read_csv("dataset/ALS.csv")
data
```

```python
# 数值型数据分布统计
plt.figure(figsize=(15,80))

numeric_cols = data.columns.tolist()[1:]  # 获取列名（除了第一个）
for i in range(len(numeric_cols)):
  plt.subplot(25, 4, i+1)  # 25行4列的子图，第i+1个
  sns.distplot(data[numeric_cols[i]].dropna())  # dropna()删除所有包含NaN的行
  plt.xlabel(numeric_cols[i])

plt.tight_layout()  # 自动调整子图参数以指定填充

plt.savefig("数值型数据分布统计.png")
plt.show()
```

```python
# 计算变量间的相关系数，观察变量间的关系
correlation = data[numeric_cols].corr()
correlation
```

```python
# 作图相关性分析
plt.figure(figsize=(20,20))

# 关闭格子线(grid line)
ax = plt.gca()
ax.grid(False)

plt.imshow(correlation, cmap='jet')
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation='vertical')
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.colorbar()
plt.tight_layout()  # tight_layout会自动调整子图参数，使之填充整个图像区域

plt.savefig("变量相关性图.png")
plt.show()
```

```python
# 删除相关性较强的特征,删除无关特征'ID','SubjectID'
col_min_max = [c for c in data.columns if c.lower()[-3:]=='max' or c.lower()[-3:]=='min' or c.lower()[-5:]=='range']
data_kmeans = data.drop(columns = col_min_max)
data_kmeans = data_kmeans.drop(columns = 'ID')
data_kmeans = data_kmeans.drop(columns = 'SubjectID')

data_kmeans
```

```python
# 查询是否有空缺值
num_of_nan = data_kmeans.isna().sum()
num_of_nan
```

```python
# Elbow方法判断聚类类簇个数
distortions = [] #用来存放设置不同簇数时的SSE值
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(data_kmeans)
    distortions.append(km.inertia_) # 获取K-means算法的SSE, #inertia_误差平方和，每个点到聚类中心点的距离之和
    
#绘制曲线
plt.plot(range(1,11),distortions,marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")

plt.savefig("判断聚类类簇个数图.png")
plt.show()
```

```python
# TSNE 聚类结果展示
tsne = TSNE().fit_transform(data_kmeans)

km = KMeans(n_clusters=4)    #y_km中保存了聚类的结果    
y_km = km.fit_predict(tsne)    #绘制不同簇的点    

plt.scatter(tsne[y_km==0,0],tsne[y_km==0,1],s=50,c="purple",marker="s",label="cluster 1")    
plt.scatter(tsne[y_km==1,0],tsne[y_km==1,1],s=50,c="green",marker="s",label="cluster 2")    
plt.scatter(tsne[y_km==2,0],tsne[y_km==2,1],s=50,c="blue",marker="o",label="cluster 3") 
plt.scatter(tsne[y_km==3,0],tsne[y_km==3,1],s=50,c="yellow",marker="o",label="cluster 4")     
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=250,marker="*",c="red",label="cluster center") #绘制簇的中心点  
plt.legend()    

plt.savefig("聚类结果图.png")
plt.show()
```
